```python
import pandas as pd
import json
import rank_bm25 as bm
from underthesea import word_tokenize
from time import time
import pickle
```

```python
import re
def process(text):    
    text = str(text)
    text = re.sub('((www\.[^\s]+)|(https?://[^\s]+))', 'URL', text)
    text = re.sub('@[^\s]+', 'AT_ABC', text)
    text = re.sub(r'\n', ' ', text)
    text = re.sub(r'([\w]+\) )', '.', text)
    text = re.sub(r'([\d]+\. )', '', text)
    text = re.sub(r'[^aAàÀảẢãÃáÁạẠăĂằẰẳẲẵẴắẮặẶâÂầẦẩẨẫẪấẤậẬbBcCdDđĐeEèÈẻẺẽẼéÉẹẸêÊềỀểỂễỄếẾệỆfFgGhHiIìÌỉỈĩĨíÍịỊjJkKlLmMnNoOòÒỏỎõÕóÓọỌôÔồỒổỔỗỖốỐộỘơƠờỜởỞỡỠớỚợỢpPqQrRsStTuUùÙủỦũŨúÚụỤưƯừỪửỬữỮứỨựỰvVwWxXyYỳỲỷỶỹỸýÝỵỴzZ0-9., ]', '', text)
    text = text.replace('.', '. ')
    text = text.replace(' .', '. ')
    text = re.sub(r'[\s]+', ' ', text)
    text = text.strip('.')
    text = text.strip('\'"')
    text = text.strip()
    text = text.lower()
    return text
```

```python
df_corpus = pd.read_csv("raw_data/final_concat_with_non_url.csv", index_col=0)
bm25_path = 'training_data/bm25_model.sav'
corpus_path = 'training_data/bm25_only_text.sav'

bm25 = pickle.load(open(bm25_path, 'rb'))
tokenized_corpus = pickle.load(open(corpus_path, 'rb'))
```

```python
df_corpus['is_non_url'] = df_corpus['is_non_url'].fillna(False)
df_corpus.to_csv("raw_data/final_concat_with_non_url.csv")
```

```python
df_corpus['combine'] = df_corpus['combine'].apply(process)
only_text = df_corpus['combine']
```

```python
df_corpus['article_id'] = df_corpus['article_id'].astype(int)
```

```python
df_corpus['law_article_combine'] = df_corpus['law_id'] + ' ' + df_corpus['article_id'].astype(str)
df_corpus['law_article_combine']
```

```python
query = process("phạt tiền không đội mũ bảo hiểm")
tokenized_query = word_tokenize(query)
docs = bm25.get_top_n(tokenized_query, only_text, n=1)
docs
```

```python
def search_2(question, n=5):
    query = process(question)
    tokenized_query = word_tokenize(query)
    docs = bm25.get_top_n(tokenized_query, only_text, n=n)
    df_search = df_corpus[df_corpus['combine'] == docs[0]]
    for i in range(1,len(docs)):
        df_search = pd.concat([df_search, df_corpus[df_corpus['combine'] == docs[i]]])
    res = []
    for i in range(len(df_search)):
        res.append(df_search.iloc[i]['law_id'] + ' ' + str(df_search.iloc[i]['article_id']))
    # print(res)
    return res
```

```python
def search_3(question, n=5):
    query = process(question)
    tokenized_query = word_tokenize(query)
    docs = bm25.get_top_n(tokenized_query, only_text, n=n)
    return docs
```
