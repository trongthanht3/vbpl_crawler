# Pre-proc function

```python
import re
def process(text):    
    text = str(text)
    text = re.sub('((www\.[^\s]+)|(https?://[^\s]+))', 'URL', text)
    text = re.sub('@[^\s]+', 'AT_ABC', text)
    text = re.sub(r'\n', ' ', text)
    text = re.sub(r'([\w]+\) )', '.', text)
    text = re.sub(r'([\d]+\. )', '', text)
    text = re.sub(r'[^aAàÀảẢãÃáÁạẠăĂằẰẳẲẵẴắẮặẶâÂầẦẩẨẫẪấẤậẬbBcCdDđĐeEèÈẻẺẽẼéÉẹẸêÊềỀểỂễỄếẾệỆfFgGhHiIìÌỉỈĩĨíÍịỊjJkKlLmMnNoOòÒỏỎõÕóÓọỌôÔồỒổỔỗỖốỐộỘơƠờỜởỞỡỠớỚợỢpPqQrRsStTuUùÙủỦũŨúÚụỤưƯừỪửỬữỮứỨựỰvVwWxXyYỳỲỷỶỹỸýÝỵỴzZ0-9., ]', '', text)
    text = text.replace('.', '. ')
    text = text.replace(' .', '. ')
    text = re.sub(r'[\s]+', ' ', text)
    text = text.strip('.')
    text = text.strip('\'"')
    text = text.strip()
    text = text.lower()
    return text
```

```python
only_text = df['combine'].apply(process)
```

```python
only_text
```

```python
# setup tokenized_corpus
import rank_bm25 as bm
from underthesea import word_tokenize
from time import time
from tqdm import tqdm

start = time()
# tokenized_corpus = [word_tokenize(doc) for doc in only_text]
tokenized_corpus = []
for i in tqdm(range(0, len(only_text))):
    tokenized_corpus.append(word_tokenize(only_text[i]))
print(time() - start)
```

```python
import pickle
filename = 'training_data/bm25_only_text.sav'
pickle.dump(tokenized_corpus, open(filename, 'wb'))

# some time later...

# load the model from disk
# tokenized_corpus = pickle.load(open(filename, 'rb'))
```

```python
# train bm25
bm25 = bm.BM25Plus(tokenized_corpus, k1=0.666, b=0.777) # plus is better than OKapi
```

```python
# dump bm25
filename = 'training_data/bm25_model.sav'
pickle.dump(bm25, open(filename, 'wb'))
```
