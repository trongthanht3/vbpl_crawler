```python
def crawl_url(page):
    start = time.time()
    url = "https://vbpl.vn/VBQPPL_UserControls/Publishing_22/TimKiem/p_KetQuaTimKiemVanBan.aspx??type=0&s=0&SearchIn=VBPQFulltext&&IsVietNamese=True&DivID=tabVB_lv1_01&Page={}&RowPerPage=1000".format(page)
#     driver.set_page_load_timeout(1000)
    driver.get(url)
#     wait = WebDriverWait(driver, 3)
    
    temp_df = pd.DataFrame(columns=["url", "lawName", "description", "expDate", "isExpire"])
    
    title = driver.find_elements(By.XPATH, '//ul[@class="listLaw"]/li')

    for tit in title:
        des = tit.find_element(By.XPATH, './/div[@class="des"]')
        law_name = tit.find_element(By.XPATH, './/p[@class="title"]')
        url = law_name.find_element(By.XPATH, './/a')
        exp_date = tit.find_elements(By.XPATH, './/p[@class="green"]')[-1]
        try:
            is_expire = tit.find_element(By.XPATH, './/p[@class="red"]').text
        except:
            pass
            is_expire = pd.NA
#         seri = pd.Series(data = {
#                             'url':url.get_attribute('href'),
#                             'law_name': law_name.text,
#                             'description': des.text
#                          },
#                          index=['url', 'law_name', 'description'])
        temp_df = temp_df.append({
                            'url':url.get_attribute('href'),
                            'lawName': law_name.text,
                            'description': des.text,
                            "expDate": exp_date.text,
                            "isExpire": is_expire
                         }, ignore_index=True)
        
    end = time.time() - start
    print("Total time:", end)
    return temp_df

# df0 = crawl_url(1)
# df0
```

```python
import re

def crawl_vbpl_each_link(url):
    driver.get(url)
    WebDriverWait(driver, 3)
    content = driver.find_element(By.XPATH, '//*[@id="toanvancontent"]')
    text = content.text
    text = text.replace("  ","")
    
    # handle article
    regex = r"^(Điều \d+\D+\n)"
    address_receive_pos = re.search(r"^(Nơi nhận:\n)", text, re.MULTILINE)
    matches = re.finditer(regex, text, re.MULTILINE)
    articles_pos = []
    articles = []    
    for matchNum, match in enumerate(matches, start=1):
        articles_pos.append((match.start(), match.end()))
    for i in range(0, len(articles_pos)-1):
        articles.append(text[articles_pos[int(i)][0]:articles_pos[int(i)+1][0]])
    
    # handle end of file
    final_pos = len(text)-1
    try:
        if int(address_receive_pos.start()) > articles_pos[-1][1]:
            final_pos = address_receive_pos.start()
    except:
        print("not found")
    articles.append(text[articles_pos[-1][0]:final_pos ])
    return articles
    
a = crawl_vbpl_each_link(df_each.iloc[20412]['url'])
print(a)
```

```python
# let's try
law_name_corner_case = ["Hiến pháp Không số",
"Bộ luật Không số",
"Bộ luật",
"Luật Số hiệu:",                        
"Luật",
"Pháp lệnh Số hiệu",                        
"Pháp lệnh",
"Lệnh ",
"Nghị quyết Nghị quyết số:",                        
"Nghị quyết liên tịch",
"Nghị quyết",                        
"Nghị định Số hiệu:",
"Nghị định", 
"van ban trung",
"văn bản trùng",
"Quyết định",
"kiểm tra hệ thống",
"Thông tư Thông tư số",
"Thông tư Trùng số ký hiệu nên hủy duyệt",
"Thông tư liên tịch", 
"(van ban nhap)",
"(trungvanban)",
"Thông tư Thông tư",
"Thông tư liên tịch",
"Thông tư",                        
"Nghị quyết",
"ngày", 
"10/7/2020",
"03/7/2020",
"lỗi",
"(văn bản mới)",
"Chỉ thị",
"Sắc lệnh", 
"Sắc luật", 
"Văn bản hợp nhất",
"Chương trình",
"Công văn"
]

def handle_law_name_by_filter(text):
    # so what we gonna do?
    # if-else of course
    for un_word in law_name_corner_case:
        if un_word == "Hiến pháp Không số":
            text = text.replace(un_word, "hpks0")
        elif un_word == "Bộ luật Không số":
            text = text.replace(un_word, "blks0")
        elif un_word == "Sắc lệnh":
            text = text.replace(un_word, "sle")
        elif un_word == "Sắc luật":
            text = text.replace(un_word, "slu")
        else:
            text = text.replace(un_word, "")
    text = text.replace(" /","/")
    text = text.replace(":", "")
    text = text.strip()
    text = text.strip(".")
    text = text.strip("`")
    text = text.strip("'")
    text = text.strip(",")
    text = text.strip("~")
    text = text.replace(" ", "/")
    text = text.lower()
    return text
```

```python
# let's try
law_name_corner_case = ["Hiến pháp Không số",
"Bộ luật Không số",
"Bộ luật",
"Luật Số hiệu:",                        
"Luật",
"Pháp lệnh Số hiệu",                        
"Pháp lệnh",
"Lệnh ",
"Nghị quyết Nghị quyết số:",                        
"Nghị quyết liên tịch",
"Nghị quyết",                        
"Nghị định Số hiệu:",
"Nghị định", 
"van ban trung",
"văn bản trùng",
"Quyết định",
"kiểm tra hệ thống",
"Thông tư Thông tư số",
"Thông tư Trùng số ký hiệu nên hủy duyệt",
"Thông tư liên tịch", 
"(van ban nhap)",
"(trungvanban)",
"Thông tư Thông tư",
"Thông tư liên tịch",
"Thông tư",                        
"Nghị quyết",
"ngày", 
"10/7/2020",
"03/7/2020",
"lỗi",
"(văn bản mới)",
"Chỉ thị",
"Sắc lệnh", 
"Sắc luật", 
"Văn bản hợp nhất",
"Chương trình",
"Công văn"
]

def handle_law_name_by_filter(text):
    # so what we gonna do?
    # if-else of course
    for un_word in law_name_corner_case:
        if un_word == "Hiến pháp Không số":
            text = text.replace(un_word, "hpks0")
        elif un_word == "Bộ luật Không số":
            text = text.replace(un_word, "blks0")
        elif un_word == "Sắc lệnh":
            text = text.replace(un_word, "sle")
        elif un_word == "Sắc luật":
            text = text.replace(un_word, "slu")
        else:
            text = text.replace(un_word, "")
    text = text.replace(" /","/")
    text = text.replace(":", "")
    text = text.strip()
    text = text.strip(".")
    text = text.strip("`")
    text = text.strip("'")
    text = text.strip(",")
    text = text.strip("~")
    text = text.replace(" ", "/")
    text = text.lower()
    return text
```
