```python
def clean_description(html_dump):
    try:
        soup = BeautifulSoup(html_dump, 'html.parser')
        strhtm = soup.prettify()
        all_link = soup.find_all('a')
        list_res = []
        for link in all_link:
            list_res.append(link.text)

        unique_list = list(dict.fromkeys(list_res))
        return unique_list
    except:
        return pd.NA
```

# Pre-process

```python
import pandas as pd
import re

# remove all html tag
def remove_html_tag(text):
    clean = re.compile('<[^>]*>')
    clean = re.sub(clean, '', text)
    return clean.strip().replace(u'\xa0', '')
```

```python
df = pd.read_csv('raw_data/raw_ThuVienPhapLuat.csv', index_col=0)
df
```

```python
for i in range(0, len(df)):
    df.at[i, 'raw_content'] = remove_html_tag(df['raw_content'][i])
```

```python
# replace duplicate \n and \r to .
def replace_duplicate(text):
    # text = text.replace('\r', '')
    return re.sub(r'[\r\n]+', '. ', text)

# # replace duplicate . except for ...
def replace_duplicate_dot(text):
    text = replace_duplicate(text)
    return re.sub(r'\.{2,}', '. ', text)

a = replace_duplicate(df['raw_content'][0])
a = replace_duplicate_dot(a)
a
```
